```python
data={"email":["Click here to win a lottery prize.",
               "Work from home ,no prior experience required.",
               "Call on this number to make payment.",
               "Loan on less interest rate,Click Here.",
               "Amazon Gift card worth Rs.500. Grab it now!.",
               "Urgent!Get your KYC done in 24 -48 hr.",
               "I phone starting from Rs.10k, limited offer, Buy now!.",
               "Get your Aadhar card verified on this portal, CLick here. From PMO Office.",
               "We processed your AppNo. 92732. youir Profile is macthed with Infosys & Genpact Openings. Click here to apply now.",
               "Start your session at DUO Lingo. Click here!",
               "A Next Skill is Just a Click Away. Simplilearn ",
               "Your Google Acoount was Logged in in a different device.",
               "Your Google Account was successfully recovered. Review now." 
              ],
      "label":["spam","spam","spam","spam","not spam","spam","spam","spam","spam","not spam","not spam","not spam","not spam"]
      }
```

```python
df=pd.DataFrame(data)
df
```

```python
def clean(text):
    text=text.lower()
    text=re.sub(r'[^\w\s]','',text)
    tokens=word_tokenize(text)
    tokens=[WordNetLemmatizer().lemmatize(word) for word in tokens]
    tokens=[word for word in tokens if word not in stopwords.words('english')]
    return "".join(tokens)
```

```python
df['email']=df['email'].apply(clean)
df
```

```python
X_train,X_test,y_train,y_test=train_test_split(df['email'],df['label'],test_size=0.2, random_state=42)
```

```python
pipeline=Pipeline([('tfidf',TfidfVectorizer()), ('Classifier', MultinomialNB())])
```

```python
pipeline.fit(X_train,y_train)
```

```python
user_email=input("Enter your email:")
cleaned_email=clean(user_email)
pred=pipeline.predict([cleaned_email])
print("Result:",pred[0])
```
